## Figuras e Eixos
Entender a diferen√ßa entre figuras e eixos √© essencial para criar e organizar gr√°ficos mais complexos.

- **Figura:** √â a janela ou √°rea onde todos os gr√°ficos s√£o desenhados. Uma figura pode conter m√∫ltiplos eixos, ou seja, podemos criar mais de um gr√°fico em uma √∫nica janela.

- **Eixo:** √â a √°rea de plotagem individual dentro de uma figura onde os dados s√£o desenhados. Cada eixo cont√©m um gr√°fico distinto.

- **Subplots:** O m√©todo `plt.subplots()` √© usado para criar uma nova figura com um ou mais eixos. Ele retorna dois objetos:
  - **fig:** Representa a figura que cont√©m todos os eixos.
  - **ax:** Refere-se ao(s) eixo(s) onde os gr√°ficos ser√£o desenhados. `ax` √© frequentemente utilizado para referenciar e modificar o conte√∫do dos eixos.


````python
# Criando uma nova figura e um eixo
fig, ax = plt.subplots()

# Mostrando a figura criada
print(f"Figura: {fig}")

# Mostrando os eixos criados
print(f"Eixos: {ax}")
````

````python
# Criando uma nova figura e um eixo
fig, ax = plt.subplots()

# Plotando o gr√°fico de linha no eixo ax
ax.plot(x, y)

# Adicionando r√≥tulo ao Eixo X
ax.set_xlabel("Eixo X")

# Adicionando r√≥tulo ao Eixo Y
ax.set_ylabel("Eixo Y")

# Definindo o t√≠tulo do gr√°fico
ax.set_title("Gr√°fico Personalizado")

# Exibindo o gr√°fico
plt.show()
````

No exemplo anterior, usamos o m√©todo `plt.subplots()` sem argumentos, o que resultou em uma figura com um √∫nico eixo, pois ele assumiu os valores padr√£o. Agora, vamos explorar mais sobre esse m√©todo.

````python
# Definindo os dados
x = ["Grupo A", "Grupo B", "Grupo C"]
y = [20, 40, 60]

# Criando uma figura com dois eixos
# Exibindo dois gr√°ficos, lado a lado, em uma √∫nica figura
fig, ax = plt.subplots(1, 2)

# Ajustando o espa√ßamento horizontal entre os gr√°ficos
fig.subplots_adjust(wspace=0.5)  # wspace = Width Space (espa√ßo horizontal)

# Plotando o gr√°fico de linha no primeiro eixo
ax[0].plot(x, y)

# Plotando o gr√°fico de barra no segundo eixo
ax[1].bar(x, y)

# Adicionando t√≠tulos aos gr√°ficos
ax[0].set_title("Gr√°fico de Linha")
ax[1].set_title("Gr√°fico de Barra")

# Exibindo os gr√°ficos
plt.show()
````

````python
# Criando duas figuras com um eixo
# Exibindo dois gr√°ficos, um abaixo do outro, em duas figuras diferentes
fig, ax = plt.subplots(2, 1)

# Ajustando o espa√ßamento vertical entre os gr√°ficos
fig.subplots_adjust(hspace=0.5)  # hspace = Height Space (espa√ßo vertical)

# Plotando o gr√°fico de linha no primeiro eixo
ax[0].plot(x, y)

# Plotando o gr√°fico de barra no segundo eixo
ax[1].bar(x, y)

# Adicionando t√≠tulos aos gr√°ficos
ax[0].set_title("Gr√°fico de Linha")
ax[1].set_title("Gr√°fico de Barra")

# Exibindo os gr√°ficos
plt.show()
````

Nos exemplos anteriores, usamos o m√©todo `plt.subplots()` para criar uma figura com dois eixos, resultando em dois gr√°ficos lado a lado. Tamb√©m criamos duas figuras separadas, cada uma com um eixo, resultando em dois gr√°ficos um abaixo do outro.

Podemos personalizar a quantidade de figuras e eixos conforme as nossas necessidades.

Mas o que acontece se tentarmos gerar dois gr√°ficos sem especificar a quantidade de figuras e eixos?

````python
# Definindo os dados
x  = ["Grupo A", "Grupo B", "Grupo C"]
y1 = [20, 40, 60]
y2 = [15, 30, 65]

# Plotando os gr√°ficos de linha
plt.plot(x, y1)
plt.plot(x, y2)

# Adicionando o t√≠tulo ao gr√°fico
plt.title("Gr√°ficos Sobrepostos")

# Exibindo os gr√°ficos
plt.show()
````

**Os gr√°ficos ficar√£o sobrepostos!** Lembre-se dessa informa√ß√£o, ela pode ser √∫til ao criar gr√°ficos de sazonalidade no futuro.

## Redimensionamento de Figuras e Ajuste de layout
Voc√™ deve ter notado que todas nossas figuras at√© agora seguiram um tamanho padr√£o (640x480 pixels). Contudo, ao trabalhar com visualiza√ß√µes de dados √© comum precisar ajustar o tamanho da figura e garantir que os subplots n√£o se sobreponham, proporcionando uma apresenta√ß√£o mais clara dos dados. Para isso, utilizamos o par√¢metro `figsize` e a fun√ß√£o `tight_layout`.

### Redimensionamento de Figuras (`figsize`)

O par√¢metro `figsize` permite definir o tamanho da figura em polegadas, ajustando a largura e a altura conforme necess√°rio. Ele √© passado como um argumento nas fun√ß√µes `plt.figure()`e `plt.subplots()`.

A sintaxe b√°sica √© `figsize=(largura, altura)`

````python
# Definindo os dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 6, 8]

# Criando uma nova figura e um eixo com tamanho 5x3 polegadas
fig, ax = plt.subplots(figsize=(5, 3))

# Plotando o gr√°fico de linha
ax.plot(x, y)

# Adicionando o t√≠tulo ao gr√°fico
ax.set_title("Gr√°fico Redimensionado")

# Exibindo o gr√°fico
plt.show()

# Mostrando a figura criada
print(f"Figura: {fig}")
````

### Ajuste de layout (`tight_layout`)

O m√©todo `tight_layout` √© extremamente √∫til para ajustar automaticamente os par√¢metros dos subplots, evitando sobreposi√ß√µes entre elementos como r√≥tulos, t√≠tulos e legendas. Isso √© especialmente importante quando trabalhamos com m√∫ltiplos subplots em uma √∫nica figura.

**Exemplo sem `tight_layout`**

````python
# Definindo os dados
x  = [1, 2, 3, 4, 5, 6]
y1 = [1, 1, 2, 2, 3, 3]
y2 = [0, 1, 0, 2, 0, 3]

# Criando duas figuras com um eixo
fig, ax = plt.subplots(2, 1)

# Plotando o gr√°fico de linha no primeiro eixo
ax[0].plot(x, y1)

# Plotando o gr√°fico de linha no segundo eixo
ax[1].plot(x, y2)

# Adicionando t√≠tulos aos gr√°ficos
ax[0].set_title("Gr√°fico de Linha 01")
ax[1].set_title("Gr√°fico de Linha 02")

# Exibindo os gr√°ficos
plt.show()
````

**Exemplo com `tight_layout`**

````python
# Criando duas figuras com um eixo
fig, ax = plt.subplots(2, 1)

# Plotando o gr√°fico de linha no primeiro eixo
ax[0].plot(x, y1)

# Plotando o gr√°fico de linha no segundo eixo
ax[1].plot(x, y2)

# Adicionando t√≠tulos aos gr√°ficos
ax[0].set_title("Gr√°fico de Linha 01")
ax[1].set_title("Gr√°fico de Linha 02")

# Ajustando o layout dos gr√°ficos
fig.tight_layout()

# Exibindo os gr√°ficos
plt.show()
````

## Limites, Margens, Escalas e Marca√ß√µes
Na visualiza√ß√£o de dados, √© crucial saber como controlar os limites dos eixos, as margens e as escalas dos gr√°ficos. Isso garante que as informa√ß√µes sejam interpretadas corretamente.

### Limites dos Eixos (`xlim` e `ylim`)

Os m√©todos `xlim` e `ylim` s√£o usados para definir os limites dos eixos X e Y, respectivamente. Permitindo escolher o intervalo de valores que ser√° exibido em cada eixo.

**Exemplo sem `xlim` e `ylim`**

````python
# Definindo os dados
x = [10, 20, 30]
y = [23, 41, 57]

# Criando uma nova figura e um eixo
fig, ax = plt.subplots()

# Plotando o gr√°fico de linha
ax.plot(x, y)

# Adicionando r√≥tulos aos eixos X e Y
ax.set_xlabel("Eixo X - Valores Padr√µes")
ax.set_ylabel("Eixo Y - Valores Padr√µes")

# Adicionando o t√≠tulo ao gr√°fico
ax.set_title("Gr√°fico de Linha - Padr√£o")

# Exibindo o gr√°fico
plt.show()
````

**Exemplo com `xlim` e `ylim`**

````python
# Criando uma nova figura e um eixo
fig, ax = plt.subplots()

# Plotando o gr√°fico de linha
ax.plot(x, y)

# Adicionando r√≥tulos aos eixos X e Y
ax.set_xlabel("Eixo X - Valores Personalizados")
ax.set_ylabel("Eixo Y - Valores Personalizados")

# Definindo o limite m√≠nimo e m√°ximo para o eixo X
ax.set_xlim(5, 35)

# Definindo o limite m√≠nimo e m√°ximo para o eixo Y
ax.set_ylim(20, 60)

# Adicionando o t√≠tulo ao gr√°fico
ax.set_title("Gr√°fico de Linha - Com Limites")

# Exibindo o gr√°fico
plt.show()
````

### Margens (`plt.margins`)
O m√©todo `plt.margins` √© utilizado para adicionar margem ao redor dos dados no gr√°fico, ajustando o espa√ßamento entre os dados e as bordas da figura. Isso √© √∫til para evitar que os dados fiquem muito pr√≥ximos das bordas do gr√°fico, proporcionando uma visualiza√ß√£o mais limpa e organizada.

**Exemplo com `plt.margins`**

````python
# Criando uma figura com tr√™s eixos, com tamanho 15x4.8 polegadas
fig, ax = plt.subplots(1, 3, figsize=(15, 4.8))

# Plotando o gr√°fico de linha no primeiro eixo
ax[0].plot(x, y)
ax[0].set_xlabel("Eixo X - Valores Padr√µes")
ax[0].set_ylabel("Eixo Y - Valores Padr√µes")
ax[0].set_title("Sem Margens")

# Plotando o gr√°fico de linha no segundo eixo
ax[1].plot(x, y)
ax[1].set_xlabel("Eixo X - Valores Redefinidos")
ax[1].set_ylabel("Eixo Y - Valores Redefinidos")
ax[1].set_title("Com Margens = 0.5")

# Adicionando margens aos eixos X e Y no segundo gr√°fico
ax[1].set_xmargin(0.5)
ax[1].set_ymargin(0.5)

# Plotando o gr√°fico de linha no terceiro eixo
ax[2].plot(x, y)
ax[2].set_xlabel("Eixo X - Valores Redefinidos")
ax[2].set_ylabel("Eixo Y - Valores Redefinidos")
ax[2].set_title("Com Margens (X = 1.5 e Y = 2)")

# Adicionando margens aos eixos X e Y no terceiro gr√°fico
ax[2].set_xmargin(1.5)
ax[2].set_ymargin(2)

# Ajustando o layout dos gr√°ficos
fig.tight_layout()

# Exibindo os gr√°ficos
plt.show()
````

### Escalas dos Eixos (`xscale` e `yscale`)
Em um gr√°fico, as escalas determinam como os valores nos eixos s√£o distribu√≠dos e apresentados, influenciando diretamente a visualiza√ß√£o e interpreta√ß√£o dos dados.

Os m√©todos `xscale` e `yscale` s√£o usados para alterar a escala dos eixos X e Y, respectivamente. O Matplotlib suporta v√°rias escalas diferentes, cada tipo de escala **possui aplica√ß√µes distintas**, dependendo do tipo de dados e das informa√ß√µes que se deseja destacar. A escolha adequada da escala pode ter um impacto significativo na compreens√£o dos dados e na clareza da apresenta√ß√£o visual.

**Importante!** Ao criar um gr√°fico sem definir a escala, o Matplotlib assumir√° automaticamente a escala linear.

Para verificar a escala adotada em um eixo espec√≠fico, utilize os comandos:

```python
ax.xaxis.get_scale()  # Verifica a escala do eixo X
ax.yaxis.get_scale()  # Verifica a escala do eixo Y
```

````python
# Desativando warnings
import warnings
warnings.simplefilter("ignore")

# Definindo os dados
# Gerando 50 valores aleat√≥rios de uma distribui√ß√£o normal com m√©dia e desvio padr√£o 0.5
# Filtrando os valores entre 0 e 1 e ordenando em ordem crescente
y = np.random.normal(loc=0.5, scale=0.5, size=50)
y = np.sort(y[(y > 0) & (y < 1)])
x = np.arange(len(y))

# Definindo a lista de escalas e t√≠tulos correspondentes
escalas = [None, "linear", "log", "logit", "symlog", "asinh"]
titulos = ["Sem escala", "Linear", "Log", "Logit", "SymmetricalLog", "Asinh"]

# Criando duas figuras com tr√™s eixos, com tamanho 15x10 polegadas
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Plotando gr√°ficos de linha com as escalas correspondentes
for i, escala in enumerate(escalas):
    linha = i // 3
    coluna = i % 3

    axs[linha, coluna].plot(x, y)
    axs[linha, coluna].set_title(titulos[i])

    # Aplicando a escala ao eixo Y, se especificada
    if escala is not None:
        axs[linha, coluna].set_yscale(escala)

# Ajustando o layout dos gr√°ficos
fig.tight_layout()

# Exibindo os gr√°ficos
plt.show()
````

### Marca√ß√µes (`Ticks`)
Os **ticks** s√£o marca√ß√µes nos eixos do gr√°fico que indicam valores de refer√™ncia. Essas marca√ß√µes podem ser num√©ricas ou categ√≥ricas e servem para destacar pontos espec√≠ficos.

Quando tra√ßamos um gr√°fico, os eixos assumem ticks padr√£o, mas podemos personaliz√°-los utilizando os m√©todos `xticks` e `yticks`. Esses m√©todos permitem definir localiza√ß√µes espec√≠ficas para as marca√ß√µes nos eixos, bem como os r√≥tulos exibidos nessas marca√ß√µes.

#### Personaliza√ß√£o
- **Definindo Localiza√ß√µes e R√≥tulos**: Utilize `plt.xticks([localiza√ß√µes], [r√≥tulos])` ou `set_xticks([localiza√ß√µes], [r√≥tulos])` para especificar onde os ticks ser√£o posicionados e quais r√≥tulos ser√£o exibidos.

- **Rota√ß√£o dos R√≥tulos**: O par√¢metro `rotation` permite girar os r√≥tulos dos ticks para melhorar a legibilidade.

- **Estilo**: Tamb√©m √© poss√≠vel personalizar a apar√™ncia dos r√≥tulos dos ticks com par√¢metros como `fontsize` e `color`, o que torna o visual do gr√°fico mais atraente. Vamos explorar isso com mais detalhes na se√ß√£o sobre **Personaliza√ß√£o de Gr√°ficos**.

A personaliza√ß√£o dos ticks contribui para a clareza e a precis√£o na comunica√ß√£o visual das informa√ß√µes apresentadas no gr√°fico.

**Exemplo sem `xticks` e `yticks`**

````python
# Definindo os dados
x = [8, 6, 4, 2, 0]
y = [1, 3, 5, 7, 9]

# Criando uma nova figura e um eixo
fig, ax = plt.subplots()

# Plotando o gr√°fico de linha no eixo ax
ax.plot(x, y)

# Adicionando r√≥tulos aos eixos X e Y
ax.set_xlabel("Eixo X")
ax.set_ylabel("Eixo Y")

# Adicionando o t√≠tulo ao gr√°fico
ax.set_title("Gr√°fico (Sem Ticks Personalizados)")

# Ajustando o layout do gr√°fico
fig.tight_layout()

# Exibindo o gr√°fico
plt.show()
````

**Exemplo com `xticks` e `yticks`**

````python
# Criando uma figura com tr√™s eixos, com tamanho 15x4.8 polegadas
fig, axs = plt.subplots(1, 3, figsize=(15, 4.8))

# Plotando o primeiro gr√°fico com ticks personalizados num√©ricos
axs[0].plot(x, y)
axs[0].set_title("Ticks Personalizados - Num√©ricos")
axs[0].set_xticks(x)  # Definindo ticks espec√≠ficos para o eixo X
axs[0].set_yticks(y)  # Definindo ticks espec√≠ficos para o eixo Y

# Plotando o segundo gr√°fico com ticks personalizados categ√≥ricos
categorias = ["A", "B", "C", "D", "E"]

axs[1].plot(x, y)
axs[1].set_title("Ticks Personalizados - Categ√≥ricos")
axs[1].set_xticks(np.sort(x))  # Definindo posi√ß√µes dos ticks no eixo X
axs[1].set_xticklabels(categorias, rotation=45) # Definindo r√≥tulos para o eixo X
axs[1].set_yticks(y)  # Definindo posi√ß√µes dos ticks no eixo Y
axs[1].set_yticklabels([f"{tick} km" for tick in y])  # Definindo r√≥tulos para o eixo Y

# Plotando o terceiro gr√°fico com ticks personalizados fixos
axs[2].plot(x, y)
axs[2].set_title("Ticks Personalizados - Valores Fixados")
axs[2].set_xticks([0, 10, 20])  # Definindo valores fixos para os ticks do eixo X
axs[2].set_yticks([0, 15, 30])  # Definindo valores fixos para os ticks do eixo Y

# Ajustando o layout dos gr√°ficos
fig.tight_layout()

# Exibindo os gr√°ficos
plt.show()
````

## Subt√≠tulos, Anota√ß√µes e Legendas
Adicionar subt√≠tulos, anota√ß√µes e legendas aos gr√°ficos √© super importante para deixar os dados mais claros e compreens√≠veis. Esses elementos ajudam a fornecer contexto, real√ßar detalhes e tornam mais f√°cil para o leitor interpretar o que est√° sendo mostrado.

### Subt√≠tulos (`plt.suptitle`)
Para adicionar um t√≠tulo centralizado √† figura, usamos a fun√ß√£o `plt.suptitle()`. Essa fun√ß√£o √© usada para colocar um subt√≠tulo geral quando voc√™ tem v√°rios subplots na mesma figura.

````python
# Definindo os dados
x = ["Grupo A", "Grupo B", "Grupo C"]
y = [20, 40, 60]

# Criando uma figura com dois eixos, com tamanho 10x4.8 polegadas
fig, ax = plt.subplots(1, 2, figsize=(10, 4.8))

# Plotando o gr√°fico de linha no primeiro eixo
ax[0].plot(x, y)
ax[0].set_title("Gr√°fico de Linha")

# Plotando o gr√°fico de barra no segundo eixo
ax[1].bar(x, y)
ax[1].set_title("Gr√°fico de Barra")

# Adicionando um subt√≠tulo √† figura
plt.suptitle("Exemplo de Subt√≠tulo")

# Ajustando o layout dos gr√°ficos
fig.tight_layout()

# Exibindo os gr√°ficos
plt.show()
````

### Anota√ß√µes (`fig.text`)
Se precisar colocar um texto em um local espec√≠fico da figura, a fun√ß√£o `fig.text()` √© a solu√ß√£o! Ela permite posicionar o texto numa posi√ß√£o espec√≠fica.

````python
# Criando uma figura com dois eixos, com tamanho 10x4.8 polegadas
fig, ax = plt.subplots(1, 2, figsize=(10, 4.8))

# Plotando o gr√°fico de linha no primeiro eixo
ax[0].plot(x, y)
ax[0].set_title("Gr√°fico de Linha")

# Plotando o gr√°fico de barra no segundo eixo
ax[1].bar(x, y)
ax[1].set_title("Gr√°fico de Barra")

# Adicionando uma anota√ß√£o √† figura usando fig.text()
fig.text(
    0.5, # Definindo a posi√ß√£o horizontal da anota√ß√£o
    0.95, # Definindo a posi√ß√£o vertical da anota√ß√£o
    "Exemplo de Anota√ß√£o com fig.text()", # Texto da anota√ß√£o
    ha="center", # Centralizando o texto horizontalmente
    fontsize=12, # Definindo o tamanho da fonte
    color="blue", # Definindo a cor do texto
)

# Ajustando o layout dos gr√°ficos
fig.tight_layout()

# Exibindo os gr√°ficos
plt.show()
````

### Anota√ß√µes (`annotate`)
A fun√ß√£o `annotate()` √© ideal para adicionar detalhes aos gr√°ficos. Com ela, voc√™ pode incluir texto e setas que direcionam a aten√ß√£o para pontos espec√≠ficos, oferecendo explica√ß√µes mais claras e contextuais.

````python
# Definindo os dados
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Criando uma nova figura e um eixo
fig, ax = plt.subplots()

# Plotando o gr√°fico de linha
ax.plot(x, y)

# Adicionando r√≥tulos aos eixos X e Y
ax.set_xlabel("Eixo X")
ax.set_ylabel("Eixo Y")

# Adicionando o t√≠tulo ao gr√°fico
ax.set_title("Gr√°fico com Anota√ß√£o")

# Adicionando uma anota√ß√£o com annotate()
ax.annotate(
    "Ponto M√°ximo", # Texto da anota√ß√£o
    xy=(5, 25), # Coordenadas do ponto a ser anotado
    xytext=(3.5, 20), # Posi√ß√£o do texto da anota√ß√£o
    arrowprops=dict(
        facecolor="red", # Definindo a cor da seta
        shrink=0.05 # Definindo o comprimento da seta
    )
)

# Exibindo o gr√°fico
plt.show()
````

### Legendas (`plt.legend`)
A fun√ß√£o `plt.legend()` serve para adicionar uma legenda no gr√°fico. Ela √© essencial para identificar e diferenciar os v√°rios elementos gr√°ficos, como linhas, barras ou outros marcadores presentes na visualiza√ß√£o dos dados.

````python
# Definindo os dados
x = ["Grupo A", "Grupo B", "Grupo C"]
y1 = [20, 40, 60]
y2 = [15, 30, 65]

# Criando uma nova figura e um eixo
fig, ax = plt.subplots()

# Plotando os gr√°ficos de linha com r√≥tulos para a legenda
ax.plot(x, y1, label="Dados 1")  # Definindo r√≥tulo da primeira linha
ax.plot(x, y2, label="Dados 2")  # Definindo r√≥tulo da segunda linha

# Adicionando o t√≠tulo ao gr√°fico
ax.set_title("Gr√°fico com Legenda A")

# Adicionando a legenda ao gr√°fico
ax.legend()

# Exibindo o gr√°fico
plt.show()
````

No gr√°fico acima, foi preciso usar uma nova propriedade ao plotar os dados: a propriedade `label`. Ela define os r√≥tulos que aparecem na legenda. Veja uma maneira alternativa de definir esses r√≥tulos abaixo:

## Personaliza√ß√£o de Gr√°ficos
Personalizar gr√°ficos √© uma habilidade essencial para qualquer analista de dados. Al√©m de deixar os gr√°ficos mais atraentes, a personaliza√ß√£o torna as informa√ß√µes mais compreens√≠veis. Com esses ajustes, voc√™ pode real√ßar pontos importantes, melhorar a legibilidade e adaptar o gr√°fico para o p√∫blico ou o contexto da apresenta√ß√£o.

### Grids

Grids s√£o linhas horizontais e verticais que atravessam o gr√°fico, facilitando a visualiza√ß√£o e interpreta√ß√£o dos dados. Elas fornecem uma refer√™ncia visual que ajuda a ler valores espec√≠ficos e a comparar diferentes pontos de dados com mais facilidade.

````python
# Definindo os dados
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 3, 5, 6, 8, 9, 11, 12]

# Criando uma nova figura e um eixo
fig, ax = plt.subplots()

# Plotando o gr√°fico de linha
ax.plot(x, y)

# Adicionando r√≥tulos aos eixos X e Y
ax.set_xlabel("Eixo X")
ax.set_ylabel("Eixo Y")

# Adicionando o t√≠tulo ao gr√°fico
ax.set_title("Gr√°fico com Grid Padr√£o")

# Exibindo o grid padr√£o
ax.grid()

# Exibindo o gr√°fico
plt.show()
````

Existem v√°rias formas de personalizar os grids de um gr√°fico. Voc√™ pode optar por exibir as linhas apenas em um dos eixos, ajustar a espessura, a cor e o estilo das linhas. As op√ß√µes s√£o bastante variadas!

````python
# Definindo os dados
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 3, 5, 6, 8, 9, 11, 12]

# Criando uma figura com quatro eixos - tamanho 15x4.8 polegadas
fig, ax = plt.subplots(1, 4, figsize=(15, 4.8))

# Grid apenas no Eixo X
ax[0].plot(x, y)
ax[0].set_xlabel("Eixo X")
ax[0].set_ylabel("Eixo Y")
ax[0].set_title("Grid apenas no Eixo X")
ax[0].grid(axis="x") # Ativando o grid apenas no eixo X

# Grid com estilo de linha
ax[1].plot(x, y)
ax[1].set_xlabel("Eixo X")
ax[1].set_ylabel("Eixo Y")
ax[1].set_title("Grid com estilo de linha")
ax[1].grid(linestyle=":") # Definindo o estilo da linha do grid como pontilhado

# Grid com linha mais espessa
ax[2].plot(x, y)
ax[2].set_xlabel("Eixo X")
ax[2].set_ylabel("Eixo Y")
ax[2].set_title("Grid com linha mais espessa")
ax[2].grid(linewidth=1.8) # Ajustando a espessura da linha do grid

# Grid com cor
ax[3].plot(x, y)
ax[3].set_xlabel("Eixo X")
ax[3].set_ylabel("Eixo Y")
ax[3].set_title("Grid com Cor")
ax[3].grid(color="Blue") # Definindo a cor do grid como azul

# Ajustando o layout
fig.tight_layout()

# Exibindo os gr√°ficos
plt.show()
````

### Spines
Spines s√£o as bordas que **delimitam a √°rea do gr√°fico**, geralmente correspondendo aos eixos X e Y. Voc√™ pode ajustar as spines para melhorar a apar√™ncia do gr√°fico.

No nosso gr√°fico, h√° quatro spines: **direita**, **esquerda**, **superior** e **inferior**.

Assim como nos grids, h√° v√°rias possibilidades com as spines: podemos remov√™-las, ou alterar suas cores, estilos e espessuras.

````python
# Definindo os dados
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 3, 5, 6, 8, 9, 11, 12]

# Criando uma figura com quatro eixos - tamanho 15x4.8 polegadas
fig, ax = plt.subplots(1, 4, figsize=(15, 4.8))

# Gr√°fico sem spines
ax[0].plot(x, y)
ax[0].set_xlabel("Eixo X")
ax[0].set_ylabel("Eixo Y")
ax[0].set_title("Gr√°fico sem spines")
# Removendo todos os spines
ax[0].spines["top"].set_visible(False)
ax[0].spines["bottom"].set_visible(False)
ax[0].spines["left"].set_visible(False)
ax[0].spines["right"].set_visible(False)

# Spines com estilo de linha
ax[1].plot(x, y)
ax[1].set_xlabel("Eixo X")
ax[1].set_ylabel("Eixo Y")
ax[1].set_title("Spines com estilo de linha")
# Definindo diferentes estilos de linha para os spines
ax[1].spines["top"].set_linestyle("--")
ax[1].spines["bottom"].set_linestyle("-.")
ax[1].spines["left"].set_linestyle(":")
ax[1].spines["right"].set_linestyle("-")

# Spines com linha mais espessa
ax[2].plot(x, y)
ax[2].set_xlabel("Eixo X")
ax[2].set_ylabel("Eixo Y")
ax[2].set_title("Spines com linha mais espessa")
# Ajustando a espessura das linhas dos spines
ax[2].spines["top"].set_linewidth(1.5)
ax[2].spines["bottom"].set_linewidth(2.5)
ax[2].spines["left"].set_linewidth(2.5)
ax[2].spines["right"].set_linewidth(1.5)

# Spines com cor
ax[3].plot(x, y)
ax[3].set_xlabel("Eixo X")
ax[3].set_ylabel("Eixo Y")
ax[3].set_title("Spines com Cor")
# Definindo cores para os spines
ax[3].spines["top"].set_color("Cyan")
ax[3].spines["bottom"].set_color("Magenta")
ax[3].spines["left"].set_color("Yellow")
ax[3].spines["right"].set_color("Gray")

# Ajustando o layout
fig.tight_layout()

# Exibindo os gr√°ficos
plt.show()
````

Mudar a cor de fundo de um gr√°fico pode transformar totalmente a maneira como o leitor o interpreta. Esse recurso √© √∫til para destacar informa√ß√µes importantes e tornar o visual mais atraente.

````python
# Definindo os dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 6, 8]

# Criando uma figura com quatro eixos - tamanho 15x4.8 polegadas
fig, ax = plt.subplots(1, 4, figsize=(15, 4.8))

# Plotando o gr√°fico de linha - Fundo: Padr√£o
ax[0].plot(x, y)

# Plotando o gr√°fico de linha - Fundo: Cinza Claro
ax[1].plot(x, y)
ax[1].set_facecolor("lightgray") # Adicionando cor de fundo

# Plotando o gr√°fico de linha - Fundo: Azul Claro
ax[2].plot(x, y)
ax[2].set_facecolor("lightskyblue") # Adicionando cor de fundo

# Plotando o gr√°fico de linha - Personalizado
ax[3].plot(x, y, color="red") # Alterando cor da linha
ax[3].set_facecolor("azure") # Adicionando cor de fundo

# Adicionando Grid Personalizado
ax[3].grid(linestyle=":", color="lightsteelblue")

# Personalizando os Spines
ax[3].spines["top"].set_visible(False)
ax[3].spines["right"].set_visible(False)
ax[3].spines["bottom"].set_color("steelblue")
ax[3].spines["left"].set_color("steelblue")

# Adicionando o t√≠tulo ao gr√°fico
ax[0].set_title("Gr√°fico - Fundo: Padr√£o")
ax[1].set_title("Gr√°fico - Fundo: Cinza Claro")
ax[2].set_title("Gr√°fico - Fundo: Azul Claro")
ax[3].set_title("Gr√°fico - Personalizado")

# Ajustando layout
fig.tight_layout()

# Exibindo o gr√°fico
plt.show()
````

### Personaliza√ß√£o de Textos

Adicionar informa√ß√µes textuais aos gr√°ficos, como vimos anteriormente, √© fundamental para melhorar a interpreta√ß√£o e apoiar o que est√° sendo exibido.

Neste guia, exploramos as propriedades `set_xlabel`, `set_title`, `fig.text`, `annotate`, entre outras. Cada uma dessas informa√ß√µes textuais pode ser personalizada, permitindo que voc√™ ajuste fontes, estilos, cores, tamanhos, espa√ßamento, e mais.

````python
# Definindo os dados
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 6, 8]

# Criando uma nova figura e um eixo
fig, ax = plt.subplots()

# Configurando a cor de fundo da figura e do eixo
fig.patch.set_facecolor("black")

# Plotando os dados
ax.plot(
    x, y, marker="o", color="red", linewidth=2,
)  # color: cor da linha, linewidth: espessura da linha

# Adicionando uma grade
ax.grid(
    color="lightgray", linestyle=":"
)  # color: cor da grade, linestyle: estilo da linha da grade

# Configurando os r√≥tulos dos eixos
ax.set_xlabel(
    "Eixo X",
    fontsize=10,  # tamanho da fonte
    color="white",  # cor do texto
    fontname="Liberation Mono",  # nome da fonte
    fontweight="bold",  # peso da fonte (negrito)
)

ax.set_ylabel(
    "Eixo Y",
    fontsize=10,  # tamanho da fonte
    color="white",  # cor do texto
    fontname="Liberation Mono",  # nome da fonte
    fontweight="bold",  # peso da fonte (negrito)
)

# Configurando as cores das bordas
ax.spines["bottom"].set_color("white")  # cor da borda inferior
ax.spines["left"].set_color("white")  # cor da borda esquerda
ax.spines["right"].set_visible(False)  # esconder a borda direita
ax.spines["top"].set_visible(False)  # esconder a borda superior

# Configurando as cores dos ticks dos eixos
ax.tick_params(axis="x", colors="gray")  # cor dos ticks do eixo X
ax.tick_params(axis="y", colors="gray")  # cor dos ticks do eixo Y

# Adicionando um t√≠tulo ao gr√°fico
ax.set_title(
    "Gr√°fico com t√≠tulo customizado",
    backgroundcolor="white",  # cor de fundo do t√≠tulo
    fontfamily="monospace",  # fam√≠lia da fonte
    fontweight="bold",  # peso da fonte (negrito)
    fontsize=12,  # tamanho da fonte
    pad=10,  # espa√ßamento entre o t√≠tulo e o gr√°fico
)

# Adicionando uma anota√ß√£o √† figura
fig.text(
    0.361,
    0.665,
    "Essa √© uma anota√ß√£o com texto customizado!\nFicou legal?",
    ha="center",  # alinhamento horizontal do texto
    color="yellow",  # cor do texto
    fontname="Liberation Sans Narrow",  # nome da fonte
    fontsize=12,  # tamanho da fonte
    fontstyle="italic",  # estilo da fonte (it√°lico)
    linespacing=1.5,  # espa√ßamento entre as linhas
    backgroundcolor="black",  # cor de fundo do texto
)

# Ajustando o layout
fig.tight_layout()

# Exibindo o gr√°fico
plt.show()
````

### Estilos (`plt.style.use`)

Utilizar `plt.style.use` no Matplotlib √© uma maneira eficiente de melhorar a apar√™ncia dos seus gr√°ficos e economizar tempo. Veja por que voc√™ deve considerar essa funcionalidade:

1. **Economia de c√≥digo**: Aplicar um estilo predefinido elimina a necessidade de configurar manualmente v√°rias propriedades de estilo para cada gr√°fico, reduzindo a quantidade de c√≥digo que voc√™ precisa escrever.

2. **Consist√™ncia visual**: Estilos predefinidos garantem uma apar√™ncia uniforme em todos os seus gr√°ficos, o que √© √≥timo para relat√≥rios e apresenta√ß√µes.

3. **Facilidade de uso**: Com `plt.style.use`, voc√™ pode alternar facilmente entre diferentes estilos para encontrar o que melhor se adapta √†s suas necessidades. Os estilos predefinidos ajudam a melhorar a est√©tica dos gr√°ficos, tornando-os mais agrad√°veis e f√°ceis de interpretar.

````python
# Obtendo todas os estilos predefinidos dispon√≠veis
print(f"Todos os estilos predefinidos dispon√≠veis na biblioteca Matplotlib:\n{plt.style.available}")
````

````python
def plotar_grafico_com_estilo(estilo):
    # Dados
    x = np.linspace(0, 10, 100)  # Gerando 100 valores igualmente espa√ßados entre 0 e 10
    y = np.sin(x)  # Calculando o seno de x

    # Aplicando o estilo escolhido
    plt.style.use(estilo)

    # Plotando o gr√°fico de linha - Aplicando estilo escolhido
    plt.plot(x, y)

    # Adicionando r√≥tulos aos eixos X e Y
    plt.xlabel("Eixo X")
    plt.ylabel("Eixo Y")

    # Adicionando o t√≠tulo ao gr√°fico
    plt.title(f"Estilo {estilo}")

    # Exibindo os gr√°ficos
    plt.show()
````

## Gr√°ficos 3D

Gr√°ficos 3D s√£o √≥timos para visualizar dados com uma dimens√£o extra. No Matplotlib, voc√™ pode criar esses gr√°ficos utilizando o m√≥dulo `mpl_toolkits.mplot3d`.

````python
# Importando a classe Axes3D para criar gr√°ficos tridimensionais
from mpl_toolkits.mplot3d import Axes3D

# Definindo os dados
x = np.linspace(-5, 5, 100)  # Gerando 100 valores igualmente espa√ßados entre -5 e 5
y = np.linspace(-5, 5, 100)  # Gerando 100 valores igualmente espa√ßados entre -5 e 5
x, y = np.meshgrid(x, y)  # Criando uma grade de valores x e y
z = np.sin(
    np.sqrt(x**2 + y**2)
)  # Calculando z como o seno da raiz quadrada da soma dos quadrados de x e y

# Criando a figura e o eixo 3D, com tamanho 10x4.8 polegadas
fig = plt.figure(figsize=(10, 4.8))

# Adicionando um eixo 3D √† figura
ax = fig.add_subplot(111, projection="3d")

# Plotando o gr√°fico de superf√≠cie 3D
surf = ax.plot_surface(
    x, y, z, cmap="ocean"
)  # Criando a superf√≠cie 3D com a paleta de cores 'ocean'

# Adicionando uma barra de cores
fig.colorbar(surf)

# Adicionando t√≠tulos e r√≥tulos aos eixos
ax.set_title("Gr√°fico de Superf√≠cie 3D")
ax.set_xlabel("Eixo X")
ax.set_ylabel("Eixo Y")
ax.set_zlabel("Eixo Z")

# Ajustando o layout
fig.tight_layout()

# Exibindo o gr√°fico
plt.show()
````

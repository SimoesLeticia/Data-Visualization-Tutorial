## Comparação com Matplotlib
Embora o Matplotlib seja uma biblioteca poderosa e amplamente utilizada para visualização de dados, o Seaborn oferece algumas vantagens significativas, especialmente para iniciantes. Aqui estão as principais vantagens de usar o Seaborn:

- **Simplicidade:** Seaborn simplifica a criação de gráficos bonitos e informativos com menos linhas de código.
- **Estética:** Os gráficos gerados pelo Seaborn têm uma aparência moderna e agradável por padrão.
- **Integração com Pandas:** Seaborn trabalha diretamente com DataFrames do Pandas, tornando mais fácil a plotagem de dados complexos.
- **Facilidade de Personalização:** Seaborn oferece uma API de alto nível para personalizar gráficos, facilitando ajustes em estilos, cores e temas.

Vamos comparar visualmente a diferença entre um histograma gerado com Matplotlib e outro gerado com Seaborn.

**Histograma com Matplotlib**

````python
# Definindo os dados
data = np.random.randn(1000)  # Gerando 1000 valores aleatórios com distribuição normal

# Plotando o histograma
plt.hist(data, bins=10)

# Adicionando o título ao gráfico
plt.title("Histograma com Matplotlib")

# Exibindo o gráfico
plt.show()
````

**Histograma com Seaborn**

````python
# Plotando o histograma
sns.histplot(data, bins=10, kde=True)

# Adicionando o título ao gráfico
plt.title("Histograma com Seaborn")

# Exibindo o gráfico
plt.show()
````

Como você pode notar, o Seaborn oferece uma forma mais simples de plotar gráficos, exigindo menos ajustes para obter visualizações esteticamente agradáveis.

## Conceitos Básicos

### Primeiro Gráfico com Seaborn
Se você já está acostumado a criar gráficos com Matplotlib, vai achar o Seaborn bem intuitivo. A principal diferença é que o Seaborn oferece uma sintaxe mais simples e várias opções de personalização por padrão. Vamos criar um gráfico de linha simples primeiro com Matplotlib e depois com Seaborn, mostrando como definir rótulos e títulos em ambos.

**Exemplo com Matplotlib**

Primeiro, vamos criar um gráfico de linha usando Matplotlib.

````python
# Criando dois arrays de dados
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 3, 5, 6, 8, 9, 11, 12]

# Gerando o gráfico de linha com plt.plot
plt.plot(x, y)

# Adicionando rótulo ao Eixo X
plt.xlabel("Eixo X")

# Adicionando rótulo ao Eixo Y
plt.ylabel("Eixo Y")

# Definindo o título do gráfico
plt.title("Meu primeiro gráfico com Matplotlib")

# Exibindo o gráfico
plt.show()
````

**Exemplo com Seaborn**

Agora, vamos criar o mesmo gráfico usando Seaborn. Note como o método `.set()` simplifica a adição de rótulos e títulos

````python
# Criando dois arrays de dados
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 3, 5, 6, 8, 9, 11, 12]

# Gerando o gráfico de linha com sns.lineplot
sns.lineplot(x=x, y=y).set(
    xlabel="Eixo X",
    ylabel="Eixo Y",
    title="Meu primeiro gráfico com Seaborn"
)

# Exibindo o gráfico
plt.show()
````

### Múltiplos Gráficos em uma Figura
Criar múltiplos gráficos em uma única figura é útil quando queremos comparar diferentes visualizações lado a lado. Em Seaborn, assim como em Matplotlib, usamos o parâmetro `ax` para especificar os eixos onde os gráficos serão desenhados. A diferença é que, em Seaborn, `ax` é passado diretamente como um argumento para as funções de plotagem, enquanto em Matplotlib os gráficos são desenhados a partir do objeto `ax`.

**Exemplo com Matplotlib**

Primeiro, vamos ver como criar múltiplos gráficos em uma figura usando Matplotlib.

````python
# Definindo os dados
x = ["Grupo A", "Grupo B", "Grupo C"]
y = [20, 40, 60]

# Criando uma figura com dois eixos - tamanho 10x4.8 polegadas
fig, ax = plt.subplots(1, 2, figsize=(10, 4.8))

# Plotando o gráfico de linha no primeiro eixo
ax[0].plot(x, y)

# Plotando o gráfico de barra no segundo eixo
ax[1].bar(x, y)

# Adicionando rótulos aos Eixos
ax[0].set_xlabel("Eixo X")
ax[0].set_ylabel("Eixo Y")
ax[1].set_xlabel("Eixo X")
ax[1].set_ylabel("Eixo Y")

# Adicionando títulos aos gráficos
ax[0].set_title("Gráfico de Linha")
ax[1].set_title("Gráfico de Barra")

# Ajustando o layout dos gráficos
fig.tight_layout()

# Exibindo os gráficos
plt.show()
````

**Exemplo com Seaborn**

Agora, vamos ver como criar os mesmos gráficos usando Seaborn. Utilizaremos o parâmetro `ax` para especificar os eixos onde queremos desenhar cada gráfico:

````python
# Definindo os dados
x = ["Grupo A", "Grupo B", "Grupo C"]
y = [20, 40, 60]

# Criando uma figura com dois eixos - tamanho 10x4.8 polegadas
fig, ax = plt.subplots(1, 2, figsize=(10, 4.8))

# Plotando o gráfico de linha no primeiro eixo
sns.lineplot(x=x, y=y, ax=ax[0]).set(
    xlabel="Eixo X",
    ylabel="Eixo Y",
    title="Gráfico de Linha"
)

# Plotando o gráfico de barra no segundo eixo
sns.barplot(x=x, y=y, ax=ax[1]).set(
    xlabel="Eixo X",
    ylabel="Eixo Y",
    title="Gráfico de Barra"
)

# Ajustando o layout dos gráficos
fig.tight_layout()

# Exibindo os gráficos
plt.show()
````

### Gráficos com Matplotlib e Seaborn
Podemos plotar gráficos usando tanto Matplotlib quanto Seaborn numa mesma figura. Veja o exemplo abaixo:

````python
# Definindo os dados
data = np.random.randn(1000)

# Criando uma figura com dois eixos - tamanho 10x4.8 polegadas
fig, ax = plt.subplots(1, 2, figsize=(10, 4.8))

# Plotando o histograma com Matplotlib
ax[0].hist(data, bins=10)

# Adicionando título e rótulos ao gráfico
ax[0].set_title("Histograma com Matplotlib")
ax[0].set_xlabel("Valores")
ax[0].set_ylabel("Contagem")

# Plotando o histograma com Seaborn
sns.histplot(data, bins=10, kde=True, ax=ax[1]).set(
    xlabel="Valores",
    ylabel="Contagem",
    title="Histograma com Seaborn"
)

# Ajustando o layout dos gráficos
fig.tight_layout()

# Exibindo o gráfico
plt.show()
````

Neste exemplo, criamos uma figura com dois gráficos lado a lado, o primeiro utilizando **Matplotlib** e o segundo **Seaborn**. Ambos os gráficos exibem um **histograma dos mesmos dados**, permitindo uma **comparação** direta **entre as duas bibliotecas**.

## Tipos de Gráficos

O Seaborn oferece diversos tipos de gráficos para explorar e visualizar dados.
Abaixo, veremos uma breve explicação de cada tipo de gráfico, juntamente com um código exemplo usando o **dataset dos pinguins**.

### Gráficos de Dispersão (`scatterplot`)
Mostram a relação entre duas variáveis contínuas. São úteis para visualizar correlações e padrões.

````python
# Criando uma figura com quatro eixos - tamanho 40x9.6 polegadas
fig, axs = plt.subplots(1, 4, figsize=(40, 9.6))

# Definindo paleta de cores personalizada para as espécies
cores_personalizadas = ["#073A61", "#DA4949", "#DD9C46"]

# Filtrando os dados para a espécie Adelie e criando um scatterplot
penguins_adelie = penguins[penguins["species"] == "Adelie"]
sns.scatterplot(
    x="flipper_length_mm",
    y="body_mass_g",
    data=penguins_adelie,
    color=cores_personalizadas[0],
    marker="s",
    ax=axs[0],
).set(
    title="Gráfico de Dispersão: Adelie\nComprimento da Nadadeira vs. Massa Corporal",
    xlabel="Comprimento da nadadeira (mm)",
    ylabel="Massa corporal (g)",
)

# Filtrando os dados para a espécie Chinstrap e criando um scatterplot
penguins_chinstrap = penguins[penguins["species"] == "Chinstrap"]
sns.scatterplot(
    x="flipper_length_mm",
    y="body_mass_g",
    data=penguins_chinstrap,
    color=cores_personalizadas[1],
    marker="h",
    ax=axs[1],
).set(
    title="Gráfico de Dispersão: Chinstrap\nComprimento da Nadadeira vs. Massa Corporal",
    xlabel="Comprimento da nadadeira (mm)",
    ylabel="Massa corporal (g)",
)

# Filtrando os dados para a espécie Gentoo e criando um scatterplot
penguins_gentoo = penguins[penguins["species"] == "Gentoo"]
sns.scatterplot(
    x="flipper_length_mm",
    y="body_mass_g",
    data=penguins_gentoo,
    color=cores_personalizadas[2],
    marker="p",
    ax=axs[2],
).set(
    title="Gráfico de Dispersão: Gentoo\nComprimento da Nadadeira vs. Massa Corporal",
    xlabel="Comprimento da nadadeira (mm)",
    ylabel="Massa corporal (g)",
)

# Criando um scatterplot para todas as espécies
sns.scatterplot(
    x="flipper_length_mm",
    y="body_mass_g",
    data=penguins,
    hue="species",
    palette=cores_personalizadas,
    marker="o",
    ax=axs[3],
).set(
    title="Gráfico de Dispersão: Todas as Espécies\nComprimento da Nadadeira vs. Massa Corporal",
    xlabel="Comprimento da nadadeira (mm)",
    ylabel="Massa corporal (g)",
)

# Ajustando layout
fig.tight_layout()

# Exibindo os gráficos
plt.show()
````

### Gráficos de Barras (barplot)
Exibem a média de uma variável contínua para diferentes categorias, facilitando a comparação de valores entre categorias.

````python
# Criando uma figura com três eixos - tamanho 40x9.6 polegadas
fig, axs = plt.subplots(1, 3, figsize=(40, 9.6))

# Definindo paleta de cores personalizada para as espécies
cores_personalizadas = ["#073A61", "#DA4949", "#DD9C46"]

# Definindo paleta de cores personalizada para o sexo
paleta_sexo_personalizada = {"Male": "#6DD2FC", "Female": "#FC6D85"}

# Criando gráfico de barras verticais
sns.barplot(
    x="species",
    y="body_mass_g",
    data=penguins,
    hue="species",
    palette=cores_personalizadas,
    ax=axs[0]
).set(
    title="Gráfico de Barras Verticais:\nEspécie vs. Massa Corporal Média",
    xlabel="Espécies",
    ylabel="Massa corporal média (g)"
)

# Criando gráfico de barras verticais agrupadas
sns.barplot(
    x="species",
    y="body_mass_g",
    hue="sex",
    data=penguins,
    palette=paleta_sexo_personalizada,
    ax=axs[1]
).set(
    title="Gráfico de Barras Verticais Agrupadas:\nEspécie e Sexo vs. Massa Corporal Média",
    xlabel="Espécies",
    ylabel="Massa corporal média (g)"
)

# Criando gráfico de barras horizontais
sns.barplot(
    x="flipper_length_mm",
    y="species",
    data=penguins,
    hue="species",
    palette=cores_personalizadas,
    ax=axs[2]
).set(
    title="Gráfico de Barras Horizontais:\nEspécie vs. Comprimento Médio da Nadadeira",
    xlabel="Comprimento da nadadeira (mm)",
    ylabel="Espécies"
)

# Ajustando layout
fig.tight_layout()

# Exibindo os gráficos
plt.show()
````

### Gráficos de Linhas (lineplot)
Usados para mostrar a evolução de uma variável ao longo do tempo ou outro eixo contínuo.

````python
# Criando uma figura com quatro eixos - tamanho 40x9.6 polegadas
fig, axs = plt.subplots(1, 4, figsize=(40, 9.6))

# Definindo paleta de cores personalizada para as espécies
cores_personalizadas = ["#073A61", "#DA4949", "#DD9C46"]

# Criando um lineplot para Massa Corporal
sns.lineplot(
    x=penguins.index,
    y="body_mass_g",
    data=penguins,
    hue="species",
    palette=cores_personalizadas,
    ax=axs[0]
).set(
    xlabel="Índice do Dataset",
    ylabel="Massa corporal (g)",
    title="Gráfico de Linhas:\nMassa Corporal ao Longo do Índice"
)

# Criando um lineplot para Comprimento da Nadadeira
sns.lineplot(
    x=penguins.index,
    y="flipper_length_mm",
    data=penguins,
    hue="species",
    palette=cores_personalizadas,
    ax=axs[1]
).set(
    xlabel="Índice do Dataset",
    ylabel="Comprimento da nadadeira (mm)",
    title="Gráfico de Linhas:\nComprimento da Nadadeira ao Longo do Índice"
)

# Criando um lineplot para Comprimento do Bico
sns.lineplot(
    x=penguins.index,
    y="bill_length_mm",
    data=penguins,
    hue="species",
    palette=cores_personalizadas,
    ax=axs[2]
).set(
    xlabel="Índice do Dataset",
    ylabel="Comprimento do bico (mm)",
    title="Gráfico de Linhas:\nComprimento do Bico ao Longo do Índice"
)

# Criando um lineplot para Profundidade do Bico
sns.lineplot(
    x=penguins.index,
    y="bill_depth_mm",
    data=penguins,
    hue="species",
    palette=cores_personalizadas,
    ax=axs[3]
).set(
    xlabel="Índice do Dataset",
    ylabel="Profundidade do bico (mm)",
    title="Gráfico de Linhas:\nProfundidade do Bico ao Longo do Índice"
)

# Ajustando layout
fig.tight_layout()

# Exibindo os gráficos
plt.show()
````

### Gráficos de Densidade (kdeplot)
Estimam a distribuição de uma variável contínua usando uma curva suavizada.

````python
# Criando uma figura com dois eixos - tamanho 40x9.6 polegadas
fig, axs = plt.subplots(1, 2, figsize=(40, 9.6))

# Definindo paleta de cores personalizada para as espécies
cores_personalizadas = ["#073A61", "#DA4949", "#DD9C46"]

# Criando um Kdeplot para Profundidade do Bico
sns.kdeplot(
    x="bill_depth_mm",
    data=penguins,
    hue="species",
    fill=True,
    palette=cores_personalizadas,
    ax=axs[0]
).set(
    xlabel="Profundidade do bico (mm)",
    ylabel="Densidade",
    title="Gráfico de Densidade: Profundidade do Bico por Espécie"
)

# Criando um Kdeplot para Comprimento do Bico
sns.kdeplot(
    x="bill_length_mm",
    data=penguins,
    hue="species",
    fill=False,
    palette=cores_personalizadas,
    ax=axs[1]
).set(
    xlabel="Comprimento do bico (mm)",
    ylabel="Densidade",
    title="Gráfico de Densidade: Comprimento do Bico por Espécie"
)

# Ajustando layout
fig.tight_layout()

# Exibindo os gráficos
plt.show()
````

### Gráficos de Caixa (boxplot)
Revelam a distribuição de uma variável contínua em diferentes categorias, incluindo mediana, quartis e outliers.

````python
# Criando uma figura com dois eixos - tamanho 40x9.6 polegadas
fig, axs = plt.subplots(1, 2, figsize=(40, 9.6))

# Definindo paleta de cores personalizada para as espécies
cores_personalizadas = ["#073A61", "#DA4949", "#DD9C46"]

# Criando um Boxplot para Profundidade do Bico
sns.boxplot(
    x="species",
    y="bill_depth_mm",
    data=penguins,
    hue="species",
    fill=True,
    palette=cores_personalizadas,
    ax=axs[0],
).set(
    xlabel="Espécies",
    ylabel="Profundidade do bico (mm)",
    title="Gráfico de Caixa: Profundidade do Bico por Espécie",
)

# Criando um Boxplot para Comprimento do Bico
sns.boxplot(
    x="species",
    y="bill_length_mm",
    data=penguins,
    hue="species",
    fill=False,
    palette=cores_personalizadas,
    ax=axs[1],
).set(
    xlabel="Espécies",
    ylabel="Comprimento do bico (mm)",
    title="Gráfico de Caixa: Comprimento do Bico por Espécie",
)

# Ajustando layout
fig.tight_layout()

# Exibindo os gráficos
plt.show()
````

### Gráficos de Violino (violinplot)

### Histogramas (histplot)
Mostram a distribuição de uma variável contínua dividida em intervalos.

````python
# Criando uma figura com quatro eixos - tamanho 40x9.6 polegadas
fig, axs = plt.subplots(1, 4, figsize=(40, 9.6))

# Definindo paleta de cores personalizada para as espécies
cores_personalizadas = ["#073A61", "#DA4949", "#DD9C46"]

# Filtrando os dados para a espécie Adelie e criando um histplot
penguins_adelie = penguins[penguins["species"] == "Adelie"]
sns.histplot(
    x="body_mass_g",
    data=penguins_adelie,
    color=cores_personalizadas[0],
    kde=True,
    ax=axs[0],
).set(
    title="Histograma\nEspécie: Adelie - Massa Corporal",
    xlabel="Massa corporal (g)",
    ylabel="Frequência",
)

# Filtrando os dados para a espécie Chinstrap e criando um scatterplot
penguins_chinstrap = penguins[penguins["species"] == "Chinstrap"]
sns.histplot(
    x="body_mass_g",
    data=penguins_chinstrap,
    color=cores_personalizadas[1],
    kde=True,
    ax=axs[1],
).set(
    title="Histograma\nEspécie: Chinstrap - Massa Corporal",
    xlabel="Massa corporal (g)",
    ylabel="Frequência",
)

# Filtrando os dados para a espécie Gentoo e criando um scatterplot
penguins_gentoo = penguins[penguins["species"] == "Gentoo"]
sns.histplot(
    x="body_mass_g",
    data=penguins_gentoo,
    color=cores_personalizadas[2],
    kde=True,
    ax=axs[2],
).set(
    title="Histograma\nEspécie: Gentoo - Massa Corporal",
    xlabel="Massa corporal (g)",
    ylabel="Frequência",
)

# Criando um scatterplot para todas as espécies
sns.histplot(
    x="body_mass_g",
    data=penguins,
    hue="species",
    palette=cores_personalizadas,
    kde=True,
    ax=axs[3],
).set(
    title="Histograma\nTodas as Espécies - Massa Corporal",
    xlabel="Massa corporal (g)",
    ylabel="Frequência",
)

# Ajustando layout
fig.tight_layout()

# Exibindo os gráficos
plt.show()
````

### Gráficos de Calor (heatmap)
Exibem uma matriz de valores em formato de tabela, onde as cores representam a magnitude dos valores.

````python
# Criando uma figura com dois eixos - tamanho 40x9.6 polegadas
fig, axs = plt.subplots(1, 2, figsize=(40, 9.6))

# Filtrando apenas as variáveis numéricas
penguins_numerico = penguins.select_dtypes(include="number")

# Calculando a matriz de correlação
corr = penguins_numerico.corr()

# Calculando o valor central (média dos valores da matriz de correlação)
central_value = corr.values.mean()

# Criando o gráfico de calor (heatmap)
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5,
    linecolor="black",
    ax=axs[0]
).set(
    title="Gráfico de Calor: Matriz de Correlação",
    xlabel="Variáveis do Dataset",
    ylabel="Variáveis do Dataset",
)

# Criando o gráfico de calor com paleta divergente
sns.heatmap(
    corr,
    cmap="coolwarm",
    center=central_value,
    linewidths=0.5,
    linecolor="white",
    ax=axs[1]
).set(
    title="Gráfico de Calor: Paleta Divergente",
    xlabel="Variáveis do Dataset",
    ylabel="Variáveis do Dataset",
)

# Ajustando layout
fig.tight_layout()

# Exibindo os gráficos
plt.show()
````

**Observação**

As *paletas de cores divergentes* são usadas para representar variáveis que sugerem oposição, com um ponto central ou neutro, permitindo que as variações ocorram em ambas as direções. São úteis em visualizações como mapas de calor, onde se deseja destacar desvios a partir de um ponto médio. Essas paletas ajudam a identificar claramente correlações negativas e positivas em torno desse ponto central.

Por exemplo:
```
Azul     = -1
Branco   =  0
Vermelho = +1
```

### Temas e Estilos
Matplotlib é uma biblioteca altamente personalizável, mas ajustar suas configurações para obter gráficos atraentes pode ser desafiador. Para facilitar esse processo, o Seaborn oferece vários **temas predefinidos** e uma interface de alto nível para controlar a **aparência dos gráficos**. Ele oferece uma **variedade de temas e estilos** que tornam os gráficos consistentemente estéticos. Além disso, permite **ajustar o contexto** dos gráficos para diferentes finalidades, como apresentações ou relatórios.

#### Temas

Seaborn permite aplicar temas globais que afetam todos os gráficos que você cria. Isso é feito usando a função `set_theme()`.

Os cinco temas predefinidos disponíveis são: `darkgrid`, `whitegrid`, `dark`, `white` e `ticks`, cada um adequado para diferentes aplicações e preferências pessoais.

````python
def plotar_grafico_com_estilo(estilo):
    # Dados
    x = np.linspace(0, 10, 100)  # Gerando 100 valores igualmente espaçados entre 0 e 10
    y = np.sin(x)  # Calculando o seno de x

    # Aplicando o estilo escolhido
    sns.set_theme(style=estilo)

    # Plotando o gráfico de linha - Aplicando estilo escolhido
    sns.lineplot(x=x, y=y).set(
        xlabel="Eixo X",
        ylabel="Eixo Y",
        title=f"Estilo: {estilo}"
    )

    # Exibindo os gráficos
    plt.show()
````

````python
# Plotando Gráfico com estilo: darkgrid
plotar_grafico_com_estilo("darkgrid")
````

````python
# Plotando Gráfico com estilo: ticks
plotar_grafico_com_estilo("ticks")
````

#### Estilos
A função `set_theme()` define o tema global para todos os gráficos, enquanto a função `set_style()` permite definir o estilo individual de cada gráfico presente em uma figura. Dessa forma, é possível ter gráficos com diferentes estilos dentro da mesma figura.

````python
def plotar_grafico_com_estilo_individual(estilo):
 # Dados
    x = np.linspace(0, 10, 100)  # Gerando 100 valores igualmente espaçados entre 0 e 10
    y = np.sin(x)  # Calculando o seno de x

    # Plotando o gráfico de linha - Aplicando estilo escolhido
    sns.lineplot(x=x, y=y).set(
        xlabel="Eixo X",
        ylabel="Eixo Y",
        title=f"Estilo: {estilo}"
    )
````

````python
# Criando uma figura com cinco eixos - tamanho 40x9.6 polegadas
fig = plt.figure(figsize=(40, 9.6))
gs = fig.add_gridspec(1, 5)

# Gerando o gráfico de linha para cada estilo
with sns.axes_style("darkgrid"):
    ax = fig.add_subplot(gs[0, 0])
    plotar_grafico_com_estilo_individual("darkgrid")

with sns.axes_style("whitegrid"):
    ax = fig.add_subplot(gs[0, 1])
    plotar_grafico_com_estilo_individual("whitegrid")

with sns.axes_style("dark"):
    ax = fig.add_subplot(gs[0, 2])
    plotar_grafico_com_estilo_individual("dark")

with sns.axes_style("white"):
    ax = fig.add_subplot(gs[0, 3])
    plotar_grafico_com_estilo_individual("white")

with sns.axes_style("ticks"):
    ax = fig.add_subplot(gs[0, 4])
    plotar_grafico_com_estilo_individual("ticks")

# Ajustando o layout dos gráficos
fig.tight_layout()

# Exibindo os gráficos
plt.show()
````

### Ajustes de Contexto
O método `sns.set_context()` é usado para definir os parâmetros que controlam o dimensionamento dos elementos do gráfico, como o tamanho dos rótulos e a espessura das linhas, sem afetar o estilo geral. Ele ajusta o `rcParams` do Matplotlib globalmente.

Existem diferentes contextos, como *notebook*, *paper*, *talk* e *poster*, que ajustam esses parâmetros conforme a necessidade, como para relatórios ou apresentações. Além disso, é possível dimensionar os elementos da fonte separadamente dos outros valores como vimos anteriormente neste guia.

````python
# Resetando para as configurações padrão
sns.reset_defaults()
````

````python
def plotar_grafico_com_contexto(contexto):
    # Dados
    x = np.linspace(0, 10, 100)  # Gerando 100 valores igualmente espaçados entre 0 e 10
    y = np.sin(x)  # Calculando o seno de x

    # Aplicando o contexto escolhido
    sns.set_context(contexto)

    # Plotando o gráfico de linha - Aplicando contexto escolhido
    sns.lineplot(x=x, y=y).set(
        xlabel="Eixo X",
        ylabel="Eixo Y",
        title=f"Contexto: {contexto}"
    )

    # Exibindo os gráficos
    plt.show()
````

### Combinação de Tema, Estilo e Contexto
Você pode combinar **temas**, **estilos** e **contextos** para criar o **visual ideal** para seus gráficos. Isso permite uma personalização completa, garantindo que seus gráficos sejam não apenas **informativos**, mas também **esteticamente agradáveis** e **adequados** ao propósito da visualização de dados.

````python
def plotar_grafico_com_personalizado(estilo, paleta, contexto):
    # Dados
    x = np.linspace(0, 10, 100)  # Gerando 100 valores igualmente espaçados entre 0 e 10
    y1 = np.sin(x)  # Calculando o seno de x
    y2 = np.cos(x)  # Calculando o cosseno de x
    y3 = np.sin(x) + np.cos(x)  # Calculando a soma do seno e do cosseno

    # Aplicando personalização escolhida
    sns.set_theme(style=estilo, palette=paleta)
    sns.set_context(contexto)

    # Criando uma figura com um eixo
    fig, ax = plt.subplots()

    # Plotando as linhas no mesmo eixo
    sns.lineplot(x=x, y=y1)
    sns.lineplot(x=x, y=y2)
    sns.lineplot(x=x, y=y3)

    # Personalizando o gráfico
    ax.set(
        xlabel="Eixo X",
        ylabel="Eixo Y",
        title=f"Estilo: {estilo}. Paleta: {paleta}. Contexto: {contexto}",
    )

    # Ajustando layout
    fig.tight_layout()

    # Exibindo o gráfico
    plt.show()
````

````python
# Plotando Gráfico com Estilo: dark, paleta: Dark2_r e contexto: paper
plotar_grafico_com_personalizado("dark", "Dark2_r", "paper")
````

````python
# Plotando Gráfico com Estilo: darkgrid, paleta: pastel e contexto: notebook
plotar_grafico_com_personalizado("darkgrid", "pastel", "notebook")
````

````python
# Plotando Gráfico com Estilo: ticks, paleta: turbo e contexto: talk
plotar_grafico_com_personalizado("ticks", "turbo", "talk")
````

````python
# Plotando Gráfico com Estilo: whitegrid, paleta: magma e contexto: poster
plotar_grafico_com_personalizado("whitegrid", "magma", "poster")
````

#### Recapitulando
- **`set_theme()`**: Define o tema global para todos os gráficos, determinando uma base visual consistente.
- **`set_style()`**: Define o estilo individual dos gráficos, permitindo que diferentes gráficos em uma mesma figura tenham aparências distintas.
- **`set_context()`**: Ajusta o contexto dos gráficos para diferentes situações, como apresentações ou relatórios, alterando o dimensionamento dos elementos para melhor se adequar ao uso pretendido.

Com essas ferramentas, é fácil personalizar seus gráficos com Seaborn para torná-los claros e visualmente atraentes, atendendo às suas necessidades específicas.
